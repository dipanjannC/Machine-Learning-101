Cumulative Distribution Functuon (CDF)



```
iris_setosa=iris[:][iris['species']=='iris_setosa']
```

```

counts,bin_edges=np.histogram(iris_setosa['petal_length'],bins=10,
                             density=True)

pdf=counts/(sum(counts))
print(pdf)
print(bin_edges)

#CDF

cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)

plt.show()
```
