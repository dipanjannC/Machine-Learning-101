```
def lines_that_equal(line_to_match, fp):
    return [line for line in fp if line == line_to_match]

def lines_that_contain(string, fp):
    return [line for line in fp if string in line]

def lines_that_start_with(string, fp):
    return [line for line in fp if line.startswith(string)]

def lines_that_end_with(string, fp):
    return [line for line in fp if line.endswith(string)]
```

```
import pandas as pd
import re

# Readging the data in txt
def read_data():
    with open("/content/BMC-ContainmentZones.js", 'r') as file:
        data = file.read().replace('\n', '')

    print('Data in Text format :' ,data,end='\n')
    return data


def get_matched_records(data):
    matches = re.findall('var marker_(.+?).addTo', data)
    print('Total Records Extracted: ', len(matches),end='\n')
    return matches 


def return_processed_data(matches):
    all_records = []
    for each in matches:
        record = []
        text = each.replace(" ","")
        id = text.split('=')[0].strip()
        # print(id)
        lat_long = re.findall(r'([^\[\]]*)', text)
        
        # print(lat_long[2].split(','))
        record.append(id)
        record.extend(lat_long[2].split(','))
        all_records.append(record)
    print('DataFrame Preview : \n')
    df = pd.DataFrame(all_records, columns = ['Marker_id', 'Latitude','Longitude'])
    return df

def main():
    data = read_data()
    return return_processed_data(get_matched_records(data))
```

```
main()
```
