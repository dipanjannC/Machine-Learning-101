```python
# US MARKETS
```

```python
us_data = pd.read_excel('/content/us_markets.xlsx')
us_data
```

```python

us_df =  pd.DataFrame()
for each in us_data.symbol:
    print(each)
    try:
        df = investpy.get_stock_historical_data(stock=each, country='united states', from_date='01/01/2019', to_date='17/11/2021').reset_index(drop=False)
        df['symbol'] = each
        us_df = us_df.append(df).reset_index(drop=True)
    except RuntimeError:
        pass
us_df
```

```python
us_final_df = us_df.merge(us_data,how='left',on='symbol')
us_final_df.to_csv('us_markets_data.csv',index = False)
```

# Indian Markets

```python
data = pd.read_excel('/content/ind_markets.xlsx')


hist_df =  pd.DataFrame()
for each in data.symbol:
    print(each)
    try:
        df = investpy.get_stock_historical_data(stock=each, country='india', from_date='01/01/2019', to_date='17/11/2021').reset_index(drop=False)
        df['symbol'] = each
        hist_df = hist_df.append(df).reset_index(drop=True)
    except RuntimeError:
        pass

final_df = hist_df.merge(data,how='left',on='symbol')
final_df.to_csv('ind_markets_data.csv',index = False)
final_df
```

#Japan Markets

#Commodities

```python
from_date = '01/01/2019'
to_date = '17/11/2021'
commodities = ['Gold','Silver','Copper','Crude Oil WTI','Natural Gas']
commodities_df =pd.DataFrame()

for each in commodities: 
    df = investpy.commodities.get_commodity_historical_data(each, from_date, to_date,country=None, as_json=False,order='ascending', interval='Daily').reset_index(drop=False)
    df['Commodity'] = each
    commodities_df = commodities_df.append(df).reset_index(drop=True)
    print(commodities_df)
```

```python
commodities_df.to_csv('commodity_prices_v1.csv',index=False)
```
