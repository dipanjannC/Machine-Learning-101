```python
# Load Data
coordinates = pd.read_csv('world_country_and_usa_states_latitude_and_longitude_values.csv')
country_coordinates = coordinates[['country_code','latitude','longitude','country']]
state_coordinates = coordinates[['usa_state_code','usa_state_latitude','usa_state_longitude','usa_state']]

df = pd.read_csv('novel-corona-virus-2019-dataset/covid_19_data.csv')
df['Country/Region'].replace(['Mainland China'], 'China',inplace=True)
df['Country/Region'].replace(['US'], 'United States',inplace=True)
df['Country'] = df['Country/Region']
# df = df[df.ObservationDate==np.max(df.ObservationDate)]
latest_updated_date = df['Last Update'].sort_values(ascending=False).head(1) # Update this line every time that you rerun the notebook
print(latest_updated_date)

# Mortality rate for every country in the dataset
df_global = df.groupby(['Country'])['Confirmed', 'Deaths', 'Recovered'].sum().reset_index()
df_global['Mortality Rate'] = np.round((df_global.Deaths.values/df_global.Confirmed.values)*100,2)
df_global = df_global.merge(country_coordinates, left_on='Country', right_on='country')
df_global = df_global[['Country','Confirmed','Deaths','Mortality Rate','Recovered','latitude','longitude','country_code']]
df_global.columns = ['Country','Confirmed','Deaths','Mortality Rate','Recovered','Latitude','Longitude','Country_Code']
df_global.to_csv('global_covid19_mortality_rates.csv')
```

```python
affected = df_global['Country'].to_list()
country_dict = {}
for country in affected:
    country_dict[country] = 0
```

```python
for text in bio_df['location'].to_list():
    for country in affected:
        if country == 'United States':
            if country in text or 'USA' in text:
                country_dict[country] += 1
        elif country == 'UK':
            if country in text or 'United Kingdom' in text:
                country_dict[country] += 1
        else:
            if country in text:
                country_dict[country] += 1
                
print(country_dict)
```

```python
country_papers = pd.DataFrame(country_dict.items(), columns=['country', 'papers'])
country_papers[country_papers['country'] == 'India']
```

```python
count = 0
for data in bio_df['location'].to_list():
    if 'India' in data:
        count += 1
print(count)
```

```python
df_global = df_global.merge(country_papers, left_on='Country', right_on='country')
df_global.to_csv('global_covid19_mortality_rates_papers.csv')
```
