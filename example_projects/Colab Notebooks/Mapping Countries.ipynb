```
import pycountry

final = []
for text in bio_df['location'].to_list():
    line=[]
    for country in pycountry.countries:
        if country.name in text:
            line.append(country.name)
        else:
           if 'USA' in text:
               line.append('USA')
           if 'UK' in text:
               line.append('UK')
           if 'UAE' in text:
               line.append('UAE')
           if 'Bolivia' in text:
               line.append('Bolivia')
           else:
               pass

    final.append(list(set(line)))
```

```
for text in bio_df['location'].to_list()[:5]:
    print(text)
```

```
len(final)
```

```
countries =[]
for each in final:
    if len(each) == 0:
        countries.append(np.nan)
    else:
        country_line = ','.join([str(elem) for elem in each]) 
        countries.append(country_line)

print(len(countries))
```

```
df_country= pd.DataFrame({'countries':countries})
df_country.head()
```
