```python
def filling_nan(*column_name):
   for c_name in column_name:
       df1 = df[c_name].fillna(method= 'ffill', inplace = True)
       return df1 
   
filling_nan('Group', 'Object')
```

```python
def pivot_to_table(data):
    df = data.copy()
    dates = pd.DataFrame(data.columns[3:],columns=['date'])
    subset = df.iloc[:,2:]
    subset = subset.T
    new_col = subset.iloc[0,:]
    subset = subset.iloc[1:,:]
    df1= subset.rename(columns=dict(zip(subset.columns[:], new_col)))
    df1 = df1.reset_index(drop = True)
    df2 = pd.concat([dates,df1],axis=1)
    return df2
```

```python
pivot_to_table(data)
```
